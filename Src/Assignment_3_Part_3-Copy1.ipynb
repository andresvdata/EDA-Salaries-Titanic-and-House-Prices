<h2>Data Cleaning
<hr>

<div class="alert alert-block alert-warning">
In this section, some attributes will be filled due to its values are important for the later analysis 

```python
# numerical
df_train.LotFrontage=df_train.groupby("Neighborhood")["LotFrontage"].transform(lambda x: x.fillna(x.median()))
df_train.MasVnrArea=df_train.MasVnrArea.fillna(0) # Those values in Nan for MasVnrArea are because this house do not have Masonry veneer.
#categorical
df_train['MSZoning'] = df_train.groupby("Neighborhood")["MSZoning"].transform(lambda x: x.fillna(x.mode()))
df_train['Electrical'] = df_train.groupby("Neighborhood")["Electrical"].transform(lambda x: x.fillna(x.mode()[0]))
```

<div class="alert alert-block alert-warning">
Since the previous code was only sustituted the columns: <b>LotFrontage</b>,<b>MSZoning</b>, <b>MasVnrArea</b> and <b>Electrical</b>,the rest of the attributes with null values can be substituted by <b>None</b> because <u>this value makes sense in the categories.

```python
missing={}
for i in df_train:
    if df_train[i].isna().sum()>0:
        missing.update({i:df_train[i].isna().sum()})
```

```python
df_missing=pd.DataFrame(missing,index=['Count']).transpose().sort_values(by='Count',ascending=False)
df_missing.columns=['missing']
for x in df_missing.index:
    df_missing.loc[x,'missing_perc']=df_missing.loc[x,'missing']/len(df_train[x])*100
```

```python
def plot_subplot(itera,n,label,compare=''):
    row=int(len(itera)/n)+len(itera)%n
    colu=len(itera)-row*n
    
    if label=='hist':
        plt.figure(figsize=(16,6*row))
        for i,j in enumerate(itera):
            i+=1
            plt.subplot(row,n,i)
            sns.violinplot(data=df_train,y=j,kde=True,hue='Neighborhood')
            plt.title('Distribution of '+j)
    elif label=='cross':
            a=0
            b=0
            fig, axes=plt.subplots(row,n,figsize=(16,8*row))
            #print(row,n)
            for i,j in enumerate(itera):
                i+=1
                my_crosstab = df_train[j].value_counts(sort=False)/sum(df_train[j].value_counts(sort=False))*100
                if row==1 or n==1:
                    #print('a',a,b)
                    ax=my_crosstab.plot(kind='bar', stacked=True, rot=90,ax=axes[a],title='Count of '+j)
                    #ax.set_xlabel(j)
                    ax.set_ylabel('% Count')
                else:
                    #print(a,b)
                    ax=my_crosstab.plot(kind='bar', stacked=True, rot=90,ax=axes[a,b],title='Count of '+j)
                    ax.set_xlabel(j)
                    ax.set_ylabel('% Count')
                if b==n-1:
                    b=0
                    a+=1
                else:
                    b+=1
```
